```python
import os
import pandas as pd
import numpy as np
from google.colab import drive
from google.colab import files
import random

drive.mount('/content/drive')
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Multilingual_Datasets/multilingual_dataset_train_final.csv') #german, japanese, turkish
dataset2 = pd.read_csv('/content/drive/MyDrive/Multilingual_Datasets/train_dataframe.csv')

dataset3 = pd.read_csv('/content/drive/MyDrive/Multilingual_Datasets/multilingual_dataset_test_final.csv')
dataset4 = pd.read_csv('/content/drive/MyDrive/Multilingual_Datasets/test_dataframe.csv')
```

```python
print(len(dataset))
print(len(dataset2))
```

```python
#change column names in dataset to 'News' and 'Label'
dataset.rename(columns={'text':'News', 'labels':'Label'}, inplace=True)
dataset3.rename(columns={'text':'News', 'labels':'Label'}, inplace=True)
```

```python
#go through labels of dataset2, change 'Fake' to 0, 'Real' to 1
dataset2.loc[dataset2['Label'] == 'Fake', "Label"] = 0
dataset2.loc[dataset2['Label'] == 'Real', "Label"] = 1

dataset4.loc[dataset4['Label'] == 'Fake', "Label"] = 0
dataset4.loc[dataset4['Label'] == 'Real', "Label"] = 1
```

```python
dataset4.value_counts
```

```python
combined_dataset_train = pd.concat([dataset, dataset2])
combined_dataset_test = pd.concat([dataset3, dataset4])
```

```python
path = '/content/drive/MyDrive/Multilingual_Datasets/merged_dataset_test.csv'
with open(path, 'w', encoding = 'utf-8-sig') as f:
  combined_dataset_test.to_csv(f)
```
