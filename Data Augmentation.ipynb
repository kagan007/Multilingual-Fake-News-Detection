```python
text1 = "Your helva nourishes my soul through its toxic meals seasoned with hopeless despair."
print(text1)
lang = translator.detect(text1).lang
translated = translator.translate(text1, src=lang, dest='tr').text
final = translator.translate(translated, src='tr', dest=lang).text
print(final)
```

```python
def backtranslate(text):
  lang = translator.detect(text).lang
  translated = translator.translate(text, src=lang, dest='su').text
  final = translator.translate(translated, src='su', dest=lang).text
  return final
```

```python
#backtranslation successful
text2 = "27 tote Tiere sollen Opfer eines natürlichen Todes geworden sein Die Hollywood-Studios Warner Bros und New Line Cinema haben Vorwürfe der Tierquälerei während der Dreharbeiten zum kommenden Blockbuster Der Hobbit: "
print(text2)
print(backtranslate(text2))
```
