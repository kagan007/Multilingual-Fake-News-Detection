```python
import os
import pandas as pd
import numpy as np
from google.colab import drive
from google.colab import files

drive.mount('/content/drive')
```

```python
data_train = pd.read_csv('/content/drive/MyDrive/Multilingual_Datasets/train_dataframe.csv')
data_test = pd.read_csv('/content/drive/MyDrive/Multilingual_Datasets/test_dataframe.csv')
```

```python
train = []
new_unit = []
haram_cells = []

for i in range(0, len(data_train)):
    if i in haram_cells:
        continue
    else:
        new_unit = [data_train['News'][i], data_train['Label'][i]]
        train.append(new_unit)
```

```python
train_df = pd.DataFrame(train, columns=['News', 'Label'])
```

```python
test = []
new_unit = []
haram_cells = []

for i in range(0, len(data_test)):
    if i in haram_cells:
        continue
    else:
        new_unit = [data_test['News'][i], data_test['Label'][i]]
        test.append(new_unit)
```

```python
test_df = pd.DataFrame(test, columns=['News', 'Label'])
```

```python
print(test_df)
```

```python
train_df.dropna(subset=['News'], inplace=True)
test_df.dropna(subset=['News'], inplace=True)
```

```python
path = '/content/drive/MyDrive/Multilingual_Datasets/train_dataframe.csv'
with open(path, 'w', encoding = 'utf-8-sig') as f:
  train_df.to_csv(f)
```

```python
path = '/content/drive/MyDrive/Multilingual_Datasets/test_dataframe.csv'
with open(path, 'w', encoding = 'utf-8-sig') as f:
  test_df.to_csv(f)
```
